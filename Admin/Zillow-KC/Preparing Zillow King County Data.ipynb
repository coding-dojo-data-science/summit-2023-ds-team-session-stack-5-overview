# Preparing Zillow King County Data 

```python
import pandas as pd
df = pd.read_csv('../../Data/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
df
```

```python
kc = df[ df['CountyName']=='King County']
kc
```

```python
id_vars = df.columns[:9]
id_vars
```

```python
kc_ts = pd.melt(kc, id_vars=id_vars, value_name="Home Value", var_name="Date" )
kc_ts
```

```python
kc_ts.to_csv('king-county-zipcode-zillow-data.csv', index=False)
```

### Minimal Preprocessing

```python
kc_ts['Date'] = pd.to_datetime(kc_ts["Date"])
kc_ts
```

```python
ts = kc_ts.set_index("Date")
ts = ts.groupby('RegionName').resample("M").asfreq()
ts
```

```python
ts_zips = ts['Home Value'].unstack(0)
ts_zips
```

```python
ts_zips.to_csv('king-county-zipcodes-as-columns-zillow-data.csv')
```
